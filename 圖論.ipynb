```python
import matplotlib.pyplot as plt

# Define the vertices and edges of the graph
vertices = ['A', 'B', 'C']
edges = [('A', 'B'), ('B', 'C'), ('A', 'C')]

# Create a dictionary to store the positions of the vertices
pos = {'A': (0, 0), 'B': (1, 1), 'C': (2, 0)}

# Draw the graph
fig, ax = plt.subplots()
for v in vertices:
    ax.scatter(pos[v][0], pos[v][1], color='red', zorder=3)
for e in edges:
    ax.plot([pos[e[0]][0], pos[e[1]][0]], [pos[e[0]][1], pos[e[1]][1]], color='black', zorder=2)

# Add labels to the vertices
for v in vertices:
    ax.annotate(v, xy=pos[v], xytext=(pos[v][0]+0.1, pos[v][1]+0.1), fontsize=12, color='blue')

# Set the limits of the axes and remove the tick marks
ax.set_xlim(-0.5, 2.5)
ax.set_ylim(-0.5, 1.5)
ax.set_xticks([])
ax.set_yticks([])

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the vertices and edges of the graph
vertices = ['A', 'B', 'C', 'D', 'E']
edges = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('A', 'C'), ('C', 'E')]

# Create a dictionary to store the positions of the vertices
pos = {'A': (0, 0), 'B': (1, 1), 'C': (2, 0), 'D': (3, 1), 'E': (4, 0)}

# Draw the graph
fig, ax = plt.subplots()
for v in vertices:
    ax.scatter(pos[v][0], pos[v][1], color='black', zorder=3)
for e in edges:
    ax.plot([pos[e[0]][0], pos[e[1]][0]], [pos[e[0]][1], pos[e[1]][1]], color='black', zorder=2)

# Add labels to the vertices
for v in vertices:
    ax.annotate(v, xy=pos[v], xytext=(pos[v][0]+0.1, pos[v][1]+0.1), fontsize=12)

# Set the limits of the axes and remove the tick marks
ax.set_xlim(-0.5, 4.5)
ax.set_ylim(-0.5, 1.5)
ax.set_xticks([])
ax.set_yticks([])

# Add the cut vertex and its edges
w_pos = (2, 1)
ax.scatter(w_pos[0], w_pos[1], color='red', zorder=3)
ax.annotate('w', xy=w_pos, xytext=(w_pos[0]+0.1, w_pos[1]+0.1), fontsize=12)
for e in [('A', 'C'), ('C', 'E')]:
    ax.plot([pos[e[0]][0], w_pos[0]], [pos[e[0]][1], w_pos[1]], color='red', zorder=1)
    ax.plot([pos[e[1]][0], w_pos[0]], [pos[e[1]][1], w_pos[1]], color='red', zorder=1)

# Add the internally disjoint paths
path1 = ['A', 'B', 'C', 'E']
path2 = ['A', 'C', 'D', 'E']
for i in range(len(path1)-1):
    ax.plot([pos[path1[i]][0], pos[path1[i+1]][0]], [pos[path1[i]][1], pos[path1[i+1]][1]], color='blue', zorder=1)
for i in range(len(path2)-1):
    ax.plot([pos[path2[i]][0], pos[path2[i+1]][0]], [pos[path2[i]][1], pos[path2[i+1]][1]], color='green', zorder=1)

# Add labels to the paths
ax.annotate('Path 1', xy=(1.5, 0.5), fontsize=12, color='blue')
ax.annotate('Path 2', xy=(2.5, 0.5), fontsize=12, color='green')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# 
nodes = ['u', 'a', 'b', 'v']

# 
edges = [('u', 'a'), ('u', 'b'), ('a', 'b'), ('a', 'v'), ('b', 'v')]

# 
fig, ax = plt.subplots()

# 
pos = {'u': (0, 1), 'a': (1, 2), 'b': (2, 0), 'v': (3, 1)}
ax.scatter([p[0] for p in pos.values()], [p[1] for p in pos.values()], s=500, color='white', edgecolors='black')
for node in nodes:
    ax.annotate(node, pos[node], ha='center', va='center')

# 
for edge in edges:
    ax.plot([pos[edge[0]][0], pos[edge[1]][0]], [pos[edge[0]][1], pos[edge[1]][1]], linewidth=2, color='black')

# 
path = [('u', 'a'), ('a', 'v')]
ax.plot([pos[p[0]][0] for p in path], [pos[p[0]][1] for p in path], linewidth=3, color='red')

#  internally disjoint 
disjoint_path = [('u', 'b'), ('b', 'v')]
ax.plot([pos[p[0]][0] for p in disjoint_path], [pos[p[0]][1] for p in disjoint_path], linewidth=3, color='blue')

# 
ax.axis('off')
ax.set_aspect('equal')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# 
G = nx.Graph()
G.add_nodes_from(['a', 'b', 'u', 'v'])

G.add_edge('a', 'b')
G.add_edge('a', 'u')
G.add_edge('a', 'v')
G.add_edge('b', 'u')
G.add_edge('b', 'v')
G.add_edge('u', 'v')

# 
pos = {'a': [0, 1], 'b': [1, 1], 'u': [0, 0], 'v': [1, 0]}
nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, {k: [v[0]+0, v[1]+0] for k, v in pos.items()})
plt.axis('off')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# 
G = nx.Graph()

# 
G.add_nodes_from(['u','v','w','x'])

# 
G.add_edges_from([('u','w'), ('v','w'), ('w','x')])

# 
pos = {'u': (0, 0), 'v': (4, 0), 'w': (2, 2), 'x': (2, -2)}
nx.draw_networkx_nodes(G, pos, node_size=500)
nx.draw_networkx_edges(G, pos)

# 
labels = {'u': 'u', 'v': 'v', 'w': 'w', 'x': 'x'}
pos_labels = {k: (v[0], v[1] + 0) for k, v in pos.items()}
nx.draw_networkx_labels(G, pos_labels, labels)

# 
path1 = [('u','w'),('w','x'),('v','w')]
path2 = [('u','w'),('w','v'),('v','x'),('x','w'),('w','u')]
nx.draw_networkx_edges(G, pos, edgelist=path1, edge_color='red', width=2)
nx.draw_networkx_edges(G, pos, edgelist=path2, edge_color='green', width=2)

plt.axis('off')
plt.show()
```
